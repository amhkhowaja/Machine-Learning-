```python
country_counts = dataframe['country'].value_counts()
country_counts
```

```python
# We can plot the countries and its frequency in the dataset
plt.figure(figsize=(10, 8))
plt.bar(country_counts.index, country_counts.values)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Histogram of Country Occurrences')
plt.xticks(rotation=45)
plt.show()
```

```python
# Lets check if there is ronaldo data in the dataset:
has_ronaldo = dataframe['player'].apply(lambda str: str.lower()).str.contains("ronaldo")

np.any(has_ronaldo) # np.any : Test whether any array element along a given axis evaluates to True.
```

```python
# As we can see that has_ronaldo value evaluates to True: which means indeed there is ronaldo.
# Now lets check the data of ronaldo
dataframe[has_ronaldo]
```

```python
# You see there are 4 results we want the information of Cristiano Ronaldo we can simply extract it like this:

dataframe[dataframe["player"] == "Cristiano Ronaldo"]
```

```python
# Let say we want to compare Lionel Messi and Cristiano Ronaldo we can create another dataframe out of the base dataframe like this

compare = pd.concat([dataframe[dataframe["player"] == "Cristiano Ronaldo"], dataframe[dataframe["player"] == "Lionel Messi"]], axis=0)
```

```python
compare
```

```python
# Lets compare them with the chart
from pandas.plotting import parallel_coordinates
plt.figure(figsize=(10, 8))

# Specify columns to include in the plot
columns_to_plot = ['height', 'weight', 'age', 'ball_control', 'dribbling','slide_tackle', 'gk_kicking']

parallel_coordinates(compare, 'player', cols=columns_to_plot)

plt.title('Player Comparison')
plt.xlabel('Attributes')
plt.ylabel('Values')
plt.xticks(rotation=45)

plt.show()
```

```python
# You can locate number of player in certain country like this :
import folium
country_counts = dataframe['country'].value_counts().reset_index()
country_counts.columns = ['country', 'count']
```

```python
# Create a map centered on a specific location (e.g., the world)
m = folium.Map(location=[0, 0], zoom_start=2)

```

```python
from geopy.geocoders import Nominatim

def get_country_coordinates(country):
    geolocator = Nominatim(user_agent="myGeocoder")
    location = geolocator.geocode(country)
    if location:
        return [location.latitude, location.longitude]
    else:
        # Return a default location (e.g., the center of the map) if the country name is not found
        return [0, 0]
```

```python
for index, row in country_counts.iterrows():
    country = row['country']
    count = row['count']
    
    # Use count to set marker properties, e.g., marker color or size
    color = 'green' if count > 2 else 'red'
    
    folium.Marker(
        location=get_country_coordinates(country),  
        popup=f"{country}: {count} players",
        icon=folium.Icon(color=color)
    ).add_to(m)
```
